```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("running-heatmap.csv", index_col=0)
data = data.loc[::-1] # reverse row order
data
```

```python
# list(data.index)
# list(data.columns)
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(
    data, 
    cmap="RdYlGn", # RdYlGn_r is reversed colors
    annot = True,
    annot_kws={"size": 12},
    fmt = ".0f",
    # vmax=200,
    )
plt.xlabel("Models / Agents", fontweight="bold", fontsize = 15)
plt.ylabel("Environments / Winds", fontweight="bold", fontsize = 15)
# plt.title("Average score over 50 episodes")
plt.savefig("heatmap.png", bbox_inches="tight")
```

```python
x = data.sum(axis=1)
x = x.sort_values(ascending = False)
```
