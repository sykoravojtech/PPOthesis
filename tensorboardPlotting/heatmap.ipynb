```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("running-heatmap.csv", index_col=0)
data = data.loc[::-1] # reverse row order
data
```

```python
# list(data.index)
# list(data.columns)
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(
    data, 
    cmap="RdYlGn", # RdYlGn_r is reversed colors
    annot = True,
    annot_kws={"size": 12},
    fmt = ".0f",
    # vmax=200,
    )
plt.xlabel("Environment / Wind", fontweight="bold", fontsize = 15)
plt.ylabel("Model / Agent", fontweight="bold", fontsize = 15)
# plt.title("Average score over 50 episodes")
plt.savefig("heatmap.png", bbox_inches="tight")
```

```python
sum_rows = data.sum(axis=1).astype('int64')
sum_rows = sum_rows.sort_values(ascending = False)
# x.to_csv("sum_table.csv")
sum_rows
```

```python
mean_rows = data.mean(axis=1).astype('int64').sort_values(ascending = False)
mean_rows
```

```python
xy = pd.concat([sum_rows, mean_rows], axis=1)
xy
```

```python
xy.columns = ['Sum of scores', 'Mean of scores']
xy
```

```python
xy.to_csv("sum_mean_table.csv")
```
